```python
%matplotlib inline
from glob import glob
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image

plt.axis('off')
sns.set()
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Palatino"],
})
```

```python
folders = {
    "mnist": ["0008", "0039", "0055"],
    "clevr": ["0037", "0068", "0061"],
    "wider": ["0009", "0020", "0091"],
    "mot15": ["0005", "0430", "0001"]
}
```

```python
fig, axes = plt.subplots(9, 9, figsize=(12,12), tight_layout=True)

dataset = "mnist"
images = folders[dataset]
for idx, image in enumerate(images):
    axes[0][0].set_title("input")
    axes[0][1].set_title("output")
    axes[0][2].set_title("with $z_\mathit{where}$")
    for model_idx, model in enumerate(("ssdir", "spair", "space")):
        row = model_idx + 3 * idx
        axes[row][0].set_ylabel(model.upper())
        path = f"{model}/{dataset}/{image}"
        input_image = f"{path}/{image}_input.png"
        output_image = f"{path}/{image}_output.png"
        boxes_image = f"{path}/{image}_with_boxes.png"
        objects_images = sorted(glob(f"{path}/{image}_obj*.png"))
        plotted = [input_image, output_image, boxes_image, *objects_images][:9]
        if len(plotted) < 9:
            plotted.extend((9 - len(plotted)) * [None])
        
        for image_idx, image_file in enumerate(plotted):
            ax = axes[row][image_idx]
            if image_file is None:
                ax.axis('off')
                continue
            ax.get_xaxis().set_ticks([])
            ax.get_yaxis().set_ticks([])
            ax.imshow(Image.open(image_file))

plt.savefig("reconstructions_mnist.png")
```

```python
import json
import numpy as np

model_dataset_mse = {}
for model in ["ssdir", "spair", "space"]:
    model_dataset_mse[model] = {}
    for dataset in ["mnist", "clevr", "wider"]:
        mses = []
        for file in glob(f"{model}/{dataset}/mse_*.txt"):
            with open(file, "r") as fp:
                mses.append(json.load(fp)["mse"])
        print(mses)
        mses = np.array(mses)
        model_dataset_mse[model][dataset] = {"mse": mses.mean(), "mse_std": mses.std()}
        
model_dataset_mse
```

```python
results = {}
for model in ("ssdir", "spair", "space"):
    results[model] = {"accuracy": [], "precision": [], "recall": [], "f1": []}

for result in glob("downstream/*.json"):
    model, seed, _ = result.split("/")[-1].split(".")[0].split("_")
    with open(result, "r") as fp:
        d = json.load(fp)
    results[model]["accuracy"].append(d["accuracy"])
    results[model]["precision"].append(d["macro avg"]["precision"])
    results[model]["recall"].append(d["macro avg"]["recall"])
    results[model]["f1"].append(d["macro avg"]["f1-score"])    
    
```

```python
for model, d in results.items():
    for metric, values in d.items():
        arr = np.array(values)
        results[model][metric] = {metric: arr.mean(), f"{metric}_std": arr.std()}
```

```python
import pandas as pd

rows = []

for key, mse in ablation_mse.items():
    grid, min_size, max_size = key.split("_")
    n_grids, min_grid, max_grid = grid.replace("grid", "")
    min_size = int(min_size.replace("min", ""))
    max_size = int(max_size.replace("max", ""))
    total_cells = sum([a * a for a in range(int(min_grid), int(max_grid) + 1)])
    avg_cell_size = int(n_grids) / total_cells
    rows.append({
        "key": key,
        "number of grids": int(n_grids),
        "min grid size": int(min_grid),
        "max grid size": int(max_grid),
        "min digit size": min_size, 
        "max digit size": max_size, 
        "MSE": mse,
        "total cells": total_cells,
        "average cell size": avg_cell_size
    })

df = pd.DataFrame(rows)
```
