```python
import sys
sys.path.append("..")
from pytorch_ssdir.modeling import SSDIR


ssdir = SSDIR.load_from_checkpoint(
    "../runs/ssdir/2aql0xlj/checkpoints/ckpt-epoch=77-val_loss=1.12.ckpt", normalize_z_present=False
)
```

```python
from tqdm import tqdm
from typing import Tuple
from pathlib import Path
import numpy as np
from PIL import Image
from albumentations import Compose, Normalize, Resize
from albumentations.pytorch import ToTensor
import torch
from pytorch_ssd.data.bboxes import iou, center_bbox_to_corner_bbox
import pyro.distributions as dist
from torchvision.ops import nms
from collections import defaultdict

with Path("/app/results/gt/MOT15-train/PETS09-S2L1/gt/gt.txt").open("r") as fp:
    gt = defaultdict(dict)
    for line in fp:
        frame_id, uid, x1, y1, w, h, *_ = line.split(",")
        gt[int(frame_id)][uid] = [float(x1), float(y1), float(x1)+float(w), float(y1)+float(h)]


def _mask_confidence_threshold(
    z_what: Tuple[torch.Tensor, torch.Tensor],
    z_where: torch.Tensor,
    z_present: torch.Tensor,
    z_depth: Tuple[torch.Tensor, torch.Tensor],
) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
    z_what_loc, z_what_scale = z_what
    z_depth_loc, z_depth_scale = z_depth

    z_what_loc = z_what_loc.view(-1, z_what_loc.shape[-1])
    z_what_scale = z_what_scale.view(-1, z_what_scale.shape[-1])
    z_where = z_where.view(-1, z_where.shape[-1])
    z_present = z_present.view(-1)
    z_depth_loc = z_depth_loc.view(-1)
    z_depth_scale = z_depth_scale.view(-1)

    mask = (
        torch.gt(z_present, 0.15)
        .nonzero(as_tuple=False)
        .squeeze(1)
    )
    z_what_loc = z_what_loc[mask]
    z_what_scale = z_what_scale[mask]
    z_what = dist.Normal(z_what_loc, z_what_scale).sample()
    z_where = z_where[mask]
    z_present = z_present[mask]
    z_depth_loc = z_depth_loc[mask]
    z_depth_scale = z_depth_scale[mask]
    z_depth = dist.Normal(z_depth_loc, z_depth_scale).sample()
    return z_what, z_where, z_present, z_depth

def _latents_nms(
    z_what: torch.Tensor,
    z_where: torch.Tensor,
    z_present: torch.Tensor,
    z_depth: torch.Tensor,
) -> Tuple[torch.Tensor, torch.Tensor, torch.Tensor, torch.Tensor]:
    z_where = center_bbox_to_corner_bbox(z_where)
    z_where[:, 0::2] *= ssdir.image_size[1]
    z_where[:, 1::2] *= ssdir.image_size[0]
    keep_mask = nms(
        boxes=z_where, scores=z_present, iou_threshold=0.45
    )
    keep_mask = keep_mask[: 100]
    return (
        z_what[keep_mask],
        z_where[keep_mask],
        z_present[keep_mask],
        z_depth[keep_mask],
    )

prepare = Compose(
    [
        Resize(*ssdir.image_size),
        Normalize(
            mean=ssdir.pixel_means,
            std=ssdir.pixel_stds,
            max_pixel_value=1.0,
        ),
        ToTensor(),
    ]
)

ssdir = ssdir.eval().cuda()

with tqdm(list(sorted(Path("/app/datasets/MOT15/train/PETS09-S2L1/img1").glob("*.jpg")))) as pbar:
    all_latents = []
    all_ids = []
    ids_per_frame = []
    for idx, image_file in enumerate(pbar, start=1):
        image = np.array(Image.open(image_file).convert("RGB")) / 255
        height, width, _ = image.shape
        prepared = prepare(image=image)["image"].cuda()
        with torch.no_grad():
            latents = ssdir.encoder(prepared.unsqueeze(0))
        z_what, z_where, z_present, z_depth = _latents_nms(
            *_mask_confidence_threshold(*latents)
        )
        z_where /= 300
        z_where[:, 0::2] *= width
        z_where[:, 1::2] *= height
        obj_ids = []
        obj_gts = []
        for key, value in gt[idx].items():
            obj_ids.append(int(key))
            obj_gts.append(value)
        for where in z_where.cpu():
            all_ids.append(obj_ids[torch.argmax(iou(torch.tensor(obj_gts), where)).item()])
        frame_objects = 0
        for latent in z_what.cpu().numpy():
            all_latents.append(latent)
            frame_objects += 1
        ids_per_frame.append(frame_objects)
            
            
        
```
